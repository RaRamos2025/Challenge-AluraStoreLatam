### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

#Calcula la facturación total por tienda
sumaTienda = sum(tienda['Precio'])
sumaTienda2 = sum(tienda2['Precio'])
sumaTienda3 = sum(tienda3['Precio'])
sumaTienda4 = sum(tienda4['Precio'])

#Imprime los resultados de la facturación total
print(f'INGRESOS TOTALES POR TIENDA\n')
print(f'Ingreso total de la tienda 1 es: ${sumaTienda:.0f}')
print(f'Ingreso total de la tienda 2 es: ${sumaTienda2:.0f}')
print(f'Ingreso total de la tienda 3 es: ${sumaTienda3:.0f}')
print(f'Ingreso total de la tienda 4 es: ${sumaTienda4:.0f}')
```

```python
import matplotlib.pyplot as plt

#Genera las listas para gráfico
tiendas = ['tienda', 'tienda2', 'tienda3', 'tienda4']
facturación = [sumaTienda, sumaTienda2, sumaTienda3, sumaTienda4]

#Gráfico de barras con facturación total por tienda
plt.bar(x=tiendas, height=facturación)
plt.title('Facturación por tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación')
plt.show()
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

#Cantidad de productos por categoría por tienda
categorias_tienda = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
categorias_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
categorias_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
categorias_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

#Imprime los resultados por categoría para cada tienda
print(f'Ventas por Categoría Tienda 1\n')
print(categorias_tienda)
print('\n')

print(f'Ventas por Categoría Tienda 2\n')
print(categorias_tienda2)
print('\n')

print(f'Ventas por Categoría Tienda 3\n')
print(categorias_tienda3)
print('\n')

print(f'Ventas por Categoría Tienda 4\n')
print(categorias_tienda4)
print('\n')
```

```python
import matplotlib.pyplot as plt

#Genera gráfico de barras horizontales por categoría para cada tienda
plt.figure(figsize=(6, 2))
barras = plt.barh(categorias_tienda.index, categorias_tienda.values)
plt.title('Ventas por Categoría Tienda')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
for barra in barras:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

plt.figure(figsize=(6, 2))
barras2 = plt.barh(categorias_tienda2.index, categorias_tienda2.values)
plt.title('Ventas por Categoría Tienda 2')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
for barra in barras2:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

plt.figure(figsize=(6, 2))
barras3 = plt.barh(categorias_tienda3.index, categorias_tienda3.values)
plt.title('Ventas por Categoría Tienda 3')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
for barra in barras3:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

plt.figure(figsize=(6, 2))
barras4 = plt.barh(categorias_tienda4.index, categorias_tienda4.values)
plt.title('Ventas por Categoría Tienda 4')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
for barra in barras4:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)
```

# 3. Calificación promedio de la tienda


```python
#Denidición del promedio de calificaciones por tienda
calif_prom = tienda['Calificación'].mean()
calif_prom2 = tienda2['Calificación'].mean()
calif_prom3 = tienda3['Calificación'].mean()
calif_prom4 = tienda4['Calificación'].mean()

# Imprime los resultados de Promedio de calificaciones por tienda, considera
# resultados con 2 decimales
print (f'Calificación promedio de la tienda 1: {calif_prom:.2f}')
print (f'Calificación promedio de la tienda 2: {calif_prom2:.2f}')
print (f'Calificación promedio de la tienda 3: {calif_prom3:.2f}')
print (f'Calificación promedio de la tienda 4: {calif_prom4:.2f}')
```

```python
#Genera gráfico de puntos con las calificaciones promedio de cada tienda
tiendas = ['tienda', 'tienda2', 'tienda3', 'tienda4']
promedios = [calif_prom, calif_prom2, calif_prom3, calif_prom4]
puntos = plt.scatter(tiendas, promedios)
plt.title('Calificación promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
plt.show()
```
